```python
import molsysmt as msm
import json
```

## Load from pharmer file

```python
with open("./data/pharmer/pharmer.json", "r") as file:
    pharmacophore = json.load(file)
```

```python
ligand = pharmacophore["ligand"]
receptor = pharmacophore["receptor"] 
```

```python
ligand_system = msm.convert(ligand, to_form="molsysmt.MolSys")
msm.info(ligand_system, target='atom')
```

```python
receptor_system = msm.convert(receptor, to_form="molsysmt.MolSys")
msm.info(receptor)
```

```python
molecular_system = msm.merge([ligand_system, receptor_system])
```

```python
msm.info(molecular_system)
```

```python
msm.extract(molecular_system, selection='molecule_type=="small_molecule"')
```

## Load pdb

```python
file = "./data/pdb/1ncr.pdb"
pdb_system = msm.convert(file)
```

```python
pdb_system = msm.remove_solvent(pdb_system)
msm.info(pdb_system, target="component")
```

```python
msm.extract(pdb_system, selection='molecule_type=="small_molecule"')
```
